```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
import pandas as pd
import matplotlib.pyplot as plt
analyser = SentimentIntensityAnalyzer()
```

```python
def print_sentiment_scores(sentence):
    snt = analyser.polarity_scores(sentence)
    print("{:-<40} {}".format(sentence, str(snt)))
    print("{:-<40} {}".format(sentence, snt['compound']))
    print("{:-<40} {}".format(sentence, snt['pos']))
    print("{:-<40} {}".format(sentence, snt['neg']))
    print("{:-<40} {}".format(sentence, snt['neu']))
```

```python
print_sentiment_scores("I have a go live date in 7 days and I have no clue how it is going to go")
```

```python
all_news = pd.read_csv('all_news.csv')
all_news = all_news.fillna('')
```

```python
# def saTitle(string):
#     snt = analyser.polarity_scores(string)
#     return snt["compound"]
    
def compound(string):
    snt = analyser.polarity_scores(string)
    return snt["compound"]

def positive(string):
    snt = analyser.polarity_scores(string)
    return snt["pos"]
    
def negative(string):
    snt = analyser.polarity_scores(string)
    return snt["neg"]

def neutral(string):
    snt = analyser.polarity_scores(string)
    return snt["neu"]
```

```python

all_news['titlePos'] = all_news.apply(lambda x: positive(x['title']), axis=1)
all_news['titleNeg'] = all_news.apply(lambda x: negative(x['title']), axis=1)
all_news['titleNeut'] = all_news.apply(lambda x: neutral(x['title']), axis=1)
all_news['titleComp'] = all_news.apply(lambda x: compound(x['title']), axis=1)

```

```python
all_news.head(20)
```

```python
all_news['descriptionPos'] = all_news.apply(lambda x: positive(x['description']), axis=1)
all_news['descriptionNeg'] = all_news.apply(lambda x: negative(x['description']), axis=1)
all_news['descriptionNeut'] = all_news.apply(lambda x: neutral(x['description']), axis=1)
```

```python
all_news['descriptionComp'] = all_news.apply(lambda x: compound(x['description']), axis=1)
```
